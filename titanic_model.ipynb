```python
data.fillna({'Age':data['Age'].mean()},inplace = True)
```

```python
data2 = pd.get_dummies(data,columns=['Sex','Embarked'],dtype = int)
```

```python
data2.head(25)
```

```python
data2['Age'] = (data2['Age']-np.min(data2['Age']))/(np.max(data2['Age']-np.min(data2['Age'])))
```

```python
data2['Fare'] = (data2['Fare']-np.min(data2['Fare']))/(np.max(data2['Fare']-np.min(data2['Fare'])))
```

```python
data3 = data2.drop(['Ticket','PassengerId'],axis=1)
```

```python
data3.head(10)
```

**Visualization**

```python
sns.boxplot(x = 'Survived',y='Age',data = data3)
plt.title('Age vs Survived')
plt.show()
```

```python
sns.countplot(x='Sex_male', hue='Survived', data=data3)
plt.title('Survival Count by Sex (Male)')
plt.show()
```

**Removing Outliers**

```python
q1 = data3['Age'].quantile(0.25)
q3 = data3['Age'].quantile(0.25)
IQR = q3 - q1
```

```python
data3 = data3[((data3['Age'] < (q1 - 1.5 * IQR)) |(data3['Age'] > (q3 + 1.5 * IQR)))]
```
